```python
# Player Statistics & Rankingsimport pandas as pdimport matplotlib.pyplot as pltimport seaborn as sns# Load datadeliveries = pd.read_csv('../data/deliveries.csv')# Top Batsmen by Runstop_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).reset_index()plt.figure(figsize=(10, 6))sns.barplot(data=top_batsmen, x='batsman_runs', y='batsman', palette='crest')plt.title('Top 10 Batsmen by Total Runs')plt.xlabel('Runs')plt.ylabel('Batsman')plt.tight_layout()plt.show()# Top Bowlers by Wickets (excluding run outs and extras)wickets = deliveries[deliveries['dismissal_kind'].isin(['bowled', 'caught', 'lbw', 'stumped', 'caught and bowled'])]top_bowlers = wickets.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10).reset_index()top_bowlers.columns = ['bowler', 'wickets']plt.figure(figsize=(10, 6))sns.barplot(data=top_bowlers, x='wickets', y='bowler', palette='flare')plt.title('Top 10 Bowlers by Wickets')plt.xlabel('Wickets')plt.ylabel('Bowler')plt.tight_layout()plt.show()# Strike Rate of Top 10 Run Scorersballs_faced = deliveries.groupby('batsman').size()runs_scored = deliveries.groupby('batsman')['batsman_runs'].sum()strike_rates = (runs_scored / balls_faced) * 100top_strike_rates = strike_rates.loc[top_batsmen['batsman']].reset_index()top_strike_rates.columns = ['batsman', 'strike_rate']plt.figure(figsize=(10, 6))sns.barplot(data=top_strike_rates, x='strike_rate', y='batsman', palette='mako')plt.title('Strike Rate of Top 10 Run Scorers')plt.xlabel('Strike Rate')plt.ylabel('Batsman')plt.tight_layout()plt.show()
```
