```python
# Head-to-Head Comparisonimport pandas as pdimport matplotlib.pyplot as pltimport seaborn as sns# Load datamatches = pd.read_csv('../data/matches.csv')# Choose two teams for comparisonteam1 = 'Mumbai Indians'team2 = 'Chennai Super Kings'# Filter head-to-head matchesh2h_matches = matches[((matches['team1'] == team1) & (matches['team2'] == team2)) |                      ((matches['team1'] == team2) & (matches['team2'] == team1))]# Win countsh2h_wins = h2h_matches['winner'].value_counts().reset_index()h2h_wins.columns = ['Team', 'Wins']plt.figure(figsize=(6, 6))sns.barplot(data=h2h_wins, x='Team', y='Wins', palette='Set2')plt.title(f'Head-to-Head: {team1} vs {team2}')plt.ylabel('Wins')plt.tight_layout()plt.show()# Match locationsvenue_counts = h2h_matches['venue'].value_counts().head(5).reset_index()venue_counts.columns = ['venue', 'matches']plt.figure(figsize=(8, 6))sns.barplot(data=venue_counts, x='matches', y='venue', palette='muted')plt.title(f'Most Common Venues: {team1} vs {team2}')plt.xlabel('Matches Played')plt.ylabel('Venue')plt.tight_layout()plt.show()print(f"Total matches between {team1} and {team2}: {len(h2h_matches)}")
```
