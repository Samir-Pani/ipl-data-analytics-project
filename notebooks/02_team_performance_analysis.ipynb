```python
# Team Performance Analysisimport pandas as pdimport matplotlib.pyplot as pltimport seaborn as sns# Load datamatches = pd.read_csv('../data/matches.csv')# Ensure date is datetimematches['date'] = pd.to_datetime(matches['date'])# Basic Win Count by Teamteam_wins = matches['winner'].value_counts().reset_index()team_wins.columns = ['Team', 'Wins']plt.figure(figsize=(10, 6))sns.barplot(data=team_wins, y='Team', x='Wins', palette='viridis')plt.title('Total Wins by Team')plt.xlabel('Wins')plt.ylabel('Team')plt.tight_layout()plt.show()# Win Trends per Seasonseason_wins = matches.groupby(['season', 'winner']).size().reset_index(name='Wins')plt.figure(figsize=(14, 6))sns.lineplot(data=season_wins, x='season', y='Wins', hue='winner', marker='o')plt.title('Team Wins per Season')plt.xticks(rotation=45)plt.tight_layout()plt.show()# Toss Win vs Match Winmatches['toss_match_win'] = matches['toss_winner'] == matches['winner']toss_win_ratio = matches['toss_match_win'].mean() * 100print(f"Teams that won the toss and match: {toss_win_ratio:.2f}%")# Pie chart for Toss vs Match Winslabels = ['Toss Win = Match Win', 'Toss Win ≠ Match Win']values = [matches['toss_match_win'].sum(), len(matches) - matches['toss_match_win'].sum()]plt.figure(figsize=(6, 6))plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'])plt.title('Impact of Toss on Match Win')plt.show()
```
