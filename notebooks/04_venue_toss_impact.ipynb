```python
# Venue & Toss Impact Studyimport pandas as pdimport matplotlib.pyplot as pltimport seaborn as sns# Load datamatches = pd.read_csv('../data/matches.csv')# Most Matches Played per Venuevenue_counts = matches['venue'].value_counts().head(10).reset_index()venue_counts.columns = ['venue', 'matches']plt.figure(figsize=(10, 6))sns.barplot(data=venue_counts, y='venue', x='matches', palette='Blues_d')plt.title('Top 10 Venues by Matches Played')plt.xlabel('Matches')plt.ylabel('Venue')plt.tight_layout()plt.show()# Toss Decision Impacttoss_decision_win = matches[matches['toss_winner'] == matches['winner']].groupby('toss_decision').size()toss_decision_total = matches.groupby('toss_decision').size()toss_success_rate = (toss_decision_win / toss_decision_total * 100).reset_index()toss_success_rate.columns = ['toss_decision', 'success_rate']plt.figure(figsize=(6, 6))sns.barplot(data=toss_success_rate, x='toss_decision', y='success_rate', palette='coolwarm')plt.title('Toss Decision Success Rate')plt.ylabel('Win % When Decision Taken')plt.tight_layout()plt.show()# Venue vs Win Percentage for Toss Winnersvenue_toss_wins = matches[matches['toss_winner'] == matches['winner']].groupby('venue').size()venue_total_matches = matches['venue'].value_counts()venue_win_percent = (venue_toss_wins / venue_total_matches * 100).dropna().sort_values(ascending=False).head(10).reset_index()venue_win_percent.columns = ['venue', 'win_percentage']plt.figure(figsize=(10, 6))sns.barplot(data=venue_win_percent, x='win_percentage', y='venue', palette='Spectral')plt.title('Top Venues Where Toss Winner Also Wins Match')plt.xlabel('Win %')plt.ylabel('Venue')plt.tight_layout()plt.show()
```
